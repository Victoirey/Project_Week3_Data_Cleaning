```python
import re
lst_reason=[]
for reason in df1.loc[:,'ReasonsForLack']:
    lst_reason.append(re.split(', (?=[A-Z])', str(reason)))
print(lst_reason)
```

```python
possible_reason=set([item for l in lst_reason for item in l])
possible_reason
```

```python
new_cols = possible_reason
df1 = df1.assign(**dict.fromkeys(new_cols, 0))
df1
```

```python
df2 = df1.copy()

for i in range(len(lst_reason)):
    for r in lst_reason[i]:
        if r in possible_reason:
            #print(i, r)
            df2.loc[i,r]=1
df2
```

```python
df2 = df2.rename(columns={'Academic Pressure':'AcademicPressure',
    'Just not interested! (Want to work in the corporate sector, or for the government or pursue research or something else)': 'NotInterested',
    'Lack of Knowledge': 'LackKnowledge', 
    'Mental Block':'MentalBlock', 
    'Not able to take a Financial Risk': 'NotAbleFinancialRisk',
    'Not willing to start a venture in India and waiting for future relocation': "FutureRelocation", 
    'Parental Pressure': 'ParentalPressure',
    'Unwillingness to take risk': 'RiskAdverse',
    'nan': 'n.p.'})
df2
```

```python
stats = df2.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
choosen_col=['Age', 'Perseverance', 'DesireToTakeInitiative',
       'Competitiveness', 'SelfReliance', 'StrongNeedToAchieve',
       'SelfConfidence', 'GoodPhysicalHealth']
```

```python
outliers=df2.loc[:, choosen_col]
```

```python
outliers
```

```python
for col in choosen_col:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 3
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = outliers[(outliers[col] < lower) | 
                   (outliers[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
```

```python
outliers
```

```python
filtered_outlier=outliers[outliers['Outlier']=='Age']
filtered_outlier
```
